```python
#creates player class to contain each players name and game history
class player:
    """a class to store the players attributes name, wins, losses, ties and total games"""
    """class default values"""
    name = "Player"
    wins = 0
    losses = 0
    ties = 0
    total_games = 0
    """Initialize class instance with name, wins, losses, total games or 0 as desired"""
    def __init__(self,name,wins,losses,ties,total_games): #initializes class instance
        self.wins = wins
        self.losses = losses
        self.ties = ties
        self.name = name
        self.total_games = total_games
        """class method to update total wins depending on win loss tie record could be used to updatet win percentage etc from in game stats used in class_updater() functiuon """
        """remember class_update() is a player class method while class_updater(player1,player2) is a function that is used to call player_update() method for both players simultaneously"""
    def player_update(self):                            ###class method to update total wins depending on win loss tie record could be used to updatet win percentage etc from in game stats
        x = self.wins + self.losses +self.ties
        self.total_games = x
        """string method to use in class_checker() function"""
    def __str__(self):                                 ###class str method so we can print player instance
        player_string = "Player's name is {} \nThey have {} wins \nThey have {} losses \nThey have {} ties \nFor a total of {} games".format(self.name,self.wins,self.losses,self.ties,self.total_games)
        return player_string
```

```python
#player class instance creator function
def player_creator():
    global player1  ###may be best to create it as nonlocal within a more general game environment?
    global player2
    player1_name = input("Player 1 enter your name:") ###sets variable to user input name
    player2_name = input("PLayer 2 enter your name:")
    player1 = player(player1_name,0,0,0,0)  ###creates player class instance named player1 with name attribute set to user string
    player2 = player(player2_name,0,0,0,0)  ###dittoo
    
#player updater function that uses class method player_update()
def player_updater(player1,player2):
    player1.player_update()
    player2.player_update()
    
#Rub it in! should be obvious what this does
def player_checker(player1,player2):
    string = "Ha! You suck {}! \n{} owns you!"
    print(player1)
    print(player2)
    if player1.wins > player2.wins:
        print(string.format(player2.name,player1.name))
    elif player2.wins > player1.wins:
        print(string.format(player1.name,player2.name))
    else:
        print("{}, {}... \nYou both suck...".format(player1.name,player2.name))
```

```python
####board functions

###prints game board from a board dictionary updates as dictionary values change
def print_game(board):
    row1 = board[7] + "|" + board[8] + "|" + board[9]
    row2 = board[4] + "|" + board[5] + "|" + board[6]
    row3 = board[1] + "|" + board[2] + "|" + board[3]
    seperator = "-+-+-"
    print(row1)
    print(seperator)
    print(row2)
    print(seperator)
    print(row3)
    
####static object that prints a board with keys image
def board_guide():
    row1 = "7" + "|" + "8" + "|" + "9"
    row2 = "4" + "|" + "5" + "|" + "6"
    row3 = "1" + "|" + "2" + "|" + "3"
    seperator = "-+-+-"
    print(row1)
    print(seperator)
    print(row2)
    print(seperator)
    print(row3)
```

```python
### function to check the 8 victory conditions using count and board as input to determine whether to return True or False
def victory_conditions(count,board):
    while count >= 5 and count < 11:                #only relevant interval where it is necessary to check for victory_conditions
        if board[7] == board[8] == board[9] != ' ':   #top row
            return True
        elif board[4] == board[5] == board[6] != ' ': #mid row
            return True
        elif board[1] == board[2] == board[3] != ' ': #bottom row
            return True
        elif board[7] == board[4] == board[1] != ' ': #left column
            return True
        elif board[8] == board[5] == board[2] != ' ': #mid column
            return True
        elif board[9] == board[6] == board[3] != ' ': #right column
            return True
        elif board[7] == board[5] == board[3] != ' ': #left to right diagnol
            return True
        elif board[9] == board[5] == board[1] != ' ': #right to left diagnol
            return True
        else:
            return False                              #return false if none above are true 
```

```python
###the game launcher function calls the player creator and/or the tic_tac_toe_game() function where the game actually takes place
def tic_tac_toe():
    print("This game may contain inappropriate language... \nUser discretion is advised...")
    decision= input("Would you like to create new players? \nEnter y for yes or any other key for no") ###too lazy to make more sophisticated how hard is it to hit y or Y anyway? 
    if decision.lower() == 'y': #if userinput = y launch player_creator () thyen launch game engine: tic_tac_toe_game()
        player_creator()
        tic_tac_toe_game()
    else:
        tic_tac_toe_game()
```

```python
### function to get user input as a string and set to the player_turn variable
#determines who starts the game as player "X" 
###returns string containing player.name attribute
def player_turnfunction():
    player_turn = input("Who is player X?")
    if player_turn == player1.name:
        player_turn = player1.name     #set player turn to appropriate player.name attribute value
    elif player_turn == player2.name:
        player_turn = player2.name
    else:
        print("How hard is it to type your own name? that you used not too long ago?")
        player_turnfunction()
    return player_turn
```

```python
#function used to restart game, launches game launcher if user reply is y or Y
def replay():
    userdecision = input("Want more shame? ENTER Y for yes Or press ANY key to abort:")
    if userdecision.lower() == "y": #user .lower() method to ensyre both Y and y are acceptable responses
          tic_tac_toe()
    else:
        print("Tired of getting whooped? So long....")
        return None
```

```python
### The actual game "engine"
def tic_tac_toe_game():
    board = {7:' ', 8:' ', 9:' ',  ###creares a dictionary with empty spaces as values to initialize empty board, keeps track of the boards current state, the game board!
             4:' ', 5:' ', 6:' ',
             1:' ', 2:' ', 3:' '}
    print("Welcome to Tic-Tac-Toe \nThe board references are below")
    board_guide()                                            #calls function that prints the board with legends to the keys
    print("To enter a move enter the number corresponding to the location on the board you want")
    player_turn = ''                                         ####initializes empty player_turn string to keep track of which player's name should appear
    player_turn = player_turnfunction() ### sets player_turn variable depending on user input to player.name
    turn = "X"                                               #initialize as "X" the variable that prints the X's and O's on the board 
    count = 1                                                ##sets count i.e the turn # or the number of iterations allowed the iterator in the while loop
    while count < 10:                                        #iterates game
        print_game(board)                                   #print board to show game state at beggining of turn
        move_string = "{} please enter your move:".format(player_turn)  ###string that asks player with name player.name to play using the player_turn variables current value
        move = int(input(move_string))                                  #ask user for input while printing move_string int from 1 to 9
        if board[move] == ' ':                                          ###check that key has empty space
            board[move] = turn                                         #write X or O on board depending on current value of variable: turn
            count +=1                                                  #increase count, the iterator and turn # by 1
        else:                                                          #if not empty space at key location print to try again and reset loop
            print("That is already filled, Try again Its literally printed for you")
            continue
        if victory_conditions(count,board) == True:                    #after move checks for victory conditions
            string = "Congratulations {} you won! \nLike the most basic game ever...".format(player_turn) #uses the player_turn variable to determine who won i.e the person who's turn just ended
            print(string)
            print_game(board)
            if player_turn == player1.name:                             ###uses value of player_turn to determine player attributes
                player1.wins += 1
                player2.losses +=1
                player_updater(player1,player2)                       ###calls player updater which invokes class method to update attributes 
                player_checker(player1,player2)                       ###prints out messages of positivity
            else:                                                      ###if not player1.name must be player2 and will update accordingly otherwise ditto
                player2.wins +=1
                player1.losses +=1
                player_updater(player1,player2)
                player_checker(player1,player2)
            replay()                                                ###launches replay() function that initializes new game or exits based on user input
            break                                                    ###breaks loop since game is over
        if turn == 'X':                                            #if turn is "X" set turn to "O"
            turn = 'O'
        else:
            turn = 'X'
        if player_turn == player1.name:                            ###updates player_turn based on player name
            player_turn = player2.name
        else:
            player_turn = player1.name
    if count == 10:                                             #not sure if I strictly need this but seems to ensure it running smoothly if loops ends without victory 
        print("It's a tie maybe give it up? like it must get boring eventually")
        print_game(board)                                     #prints final board
        player1.ties +=1                                          #updates player attributes                                           
        player2.ties +=1
        player_updater(player1,player2)
        player_checker(player1,player2)                       ##checks players stats and trash talks
        replay()                                                #replay function again could put outside loop but was finicky earlier
    return True                                             #function is game engine so return True if works could be useful to use in a future auto_debugger
```
